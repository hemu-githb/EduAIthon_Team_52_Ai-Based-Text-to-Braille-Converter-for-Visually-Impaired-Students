```python
!pip install gTTS googletrans==4.0.0-rc1 pytesseract


from google.colab import files
from PIL import Image
import pytesseract
from googletrans import Translator
from gtts import gTTS
from IPython.display import Audio, display

BRAILLE_MAPPINGS = {
    'en': {
        'a': '‚†Å', 'b': '‚†É', 'c': '‚†â', 'd': '‚†ô', 'e': '‚†ë',
        'f': '‚†ã', 'g': '‚†õ', 'h': '‚†ì', 'i': '‚†ä', 'j': '‚†ö',
        'k': '‚†Ö', 'l': '‚†á', 'm': '‚†ç', 'n': '‚†ù', 'o': '‚†ï',
        'p': '‚†è', 'q': '‚†ü', 'r': '‚†ó', 's': '‚†é', 't': '‚†û',
        'u': '‚†•', 'v': '‚†ß', 'w': '‚†∫', 'x': '‚†≠', 'y': '‚†Ω',
        'z': '‚†µ',
        'A': '‚†Å', 'B': '‚†É', 'C': '‚†â', 'D': '‚†ô', 'E': '‚†ë',
        'F': '‚†ã', 'G': '‚†õ', 'H': '‚†ì', 'I': '‚†ä', 'J': '‚†ö',
        'K': '‚†Ö', 'L': '‚†á', 'M': '‚†ç', 'N': '‚†ù', 'O': '‚†ï',
        'P': '‚†è', 'Q': '‚†ü', 'R': '‚†ó', 'S': '‚†é', 'T': '‚†û',
        'U': '‚†•', 'V': '‚†ß', 'W': '‚†∫', 'X': '‚†≠', 'Y': '‚†Ω',
        'Z': '‚†µ',
        ' ': ' ', '.': '‚†≤', ',': '‚†Ç', '?': '‚†¶', '!': '‚†ñ',
        '-': '‚†§', "'": '‚†Ñ', '"': '‚†∂', ':': '‚†í', ';': '‚†Ü',
        '0': '‚†¥', '1': '‚†Ç', '2': '‚†Ü', '3': '‚†í', '4': '‚†≤',
        '5': '‚†¢', '6': '‚†ñ', '7': '‚†∂', '8': '‚†¶', '9': '‚†î'
    },
    'kn': {
        '‡≤Ö': '‚†Å', '‡≤Ü': '‚†É', '‡≤á': '‚†â', '‡≤à': '‚†ã', '‡≤â': '‚†ô',
        '‡≤ä': '‚†ë', '‡≤ã': '‚†ö', '‡≤é': '‚†õ', '‡≤è': '‚†ì', '‡≤ê': '‚†ä',
        '‡≤í': '‚†ö', '‡≤ì': '‚†Ö', '‡≤î': '‚†á', '‡≤ï': '‚†à', '‡≤ñ': '‚†í',
        '‡≤ó': '‚†å', '‡≤ò': '‚†ú', '‡≤ô': '‚†§', '‡≤ö': '‚†®', '‡≤ú': '‚†¥',
        '‡≤ù': '‚†ú', '‡≤ü': '‚†§', '‡≤°': '‚†•', '‡≤£': '‚†±', '‡≤§': '‚†ê',
        '‡≤¶': '‚†ö', '‡≤®': '‚†ù', '‡≤™': '‚†ï', '‡≤¨': '‚†É', '‡≤Æ': '‚†é',
        '‡≤Ø': '‚†Ω', '‡≤∞': '‚†ó', '‡≤≤': '‚†á', '‡≤µ': '‚†∫', '‡≤∂': '‚†π',
        '‡≤∏': '‚†Ø', '‡≤π': '‚†ì', '‡≤≥': '‚†™', ' ': ' ', '.': '‚†≤'
    },
    'hi': {
        '‡§Ö': '‚†Å', '‡§Ü': '‚†É', '‡§á': '‚†â', '‡§à': '‚†ã', '‡§â': '‚†ô',
        '‡§ä': '‚†ë', '‡§ã': '‚†ö', '‡§è': '‚†õ', '‡§ê': '‚†ì', '‡§ì': '‚†ä',
        '‡§î': '‚†Ö', '‡§ï': '‚†à', '‡§ñ': '‚†í', '‡§ó': '‚†å', '‡§ò': '‚†ú',
        '‡§ö': '‚†®', '‡§ú': '‚†¥', '‡§ù': '‚†ú', '‡§ü': '‚†§', '‡§°': '‚†•',
        '‡§£': '‚†±', '‡§§': '‚†ê', '‡§¶': '‚†ö', '‡§®': '‚†ù', '‡§™': '‚†ï',
        '‡§¨': '‚†É', '‡§Æ': '‚†é', '‡§Ø': '‚†Ω', '‡§∞': '‚†ó', '‡§≤': '‚†á',
        '‡§µ': '‚†∫', '‡§∂': '‚†π', '‡§∏': '‚†Ø', '‡§π': '‚†ì', ' ': ' ', '.': '‚†≤'
    },
    'te': {
        '‡∞Ö': '‚†Å', '‡∞Ü': '‚†É', '‡∞á': '‚†â', '‡∞à': '‚†ã', '‡∞â': '‚†ô',
        '‡∞ä': '‚†ë', '‡∞ã': '‚†ö', '‡∞é': '‚†õ', '‡∞è': '‚†ì', '‡∞ê': '‚†ä',
        '‡∞í': '‚†ö', '‡∞ì': '‚†Ö', '‡∞î': '‚†á', '‡∞ï': '‚†à', '‡∞ñ': '‚†í',
        '‡∞ó': '‚†å', '‡∞ò': '‚†ú', '‡∞ö': '‚†®', '‡∞ú': '‚†¥', '‡∞ü': '‚†§',
        '‡∞°': '‚†•', '‡∞£': '‚†±', '‡∞§': '‚†ê', '‡∞¶': '‚†ö', '‡∞®': '‚†ù',
        '‡∞™': '‚†ï', '‡∞¨': '‚†É', '‡∞Æ': '‚†é', '‡∞Ø': '‚†Ω', '‡∞∞': '‚†ó',
        '‡∞≤': '‚†á', '‡∞µ': '‚†∫', '‡∞∂': '‚†π', '‡∞∏': '‚†Ø', '‡∞π': '‚†ì',
        '‡∞≥': '‚†™', ' ': ' ', '.': '‚†≤'
    }
}

def translate_text(text, target_lang):
    translator = Translator()
    try:
        translated = translator.translate(text, src='en', dest=target_lang)
        return translated.text
    except Exception as e:
        return f"Translation error: {e}"

def convert_to_braille(text, lang_code):
    mapping = BRAILLE_MAPPINGS.get(lang_code, {})
    return ''.join(mapping.get(char, '?') for char in text)

def beautify_braille_output(braille_text, line_width=30):
    clean_text = braille_text.replace('\n', '').replace('\r', '')
    lines = [clean_text[i:i+line_width] for i in range(0, len(clean_text), line_width)]
    border = '‚†ø' + '‚†ø' * line_width + '‚†ø'
    formatted_output = [border]
    for line in lines:
        padded = line + '‚†Ä' * (line_width - len(line))  # Braille blank for padding
        formatted_output.append('‚†ø' + padded + '‚†ø')
    formatted_output.append(border)
    return '\n'.join(formatted_output)

def speak_text(text, lang_code):
    tts = gTTS(text=text, lang=lang_code)
    tts.save("output.mp3")
    return Audio("output.mp3", autoplay=True)

print("üåê Choose a target language:\n1. Kannada\n2. Hindi\n3. Telugu\n4. Just Braille output")
lang_choice = input("Enter choice (1/2/3/4): ").strip()
lang_map = {'1': 'kn', '2': 'hi', '3': 'te', '4': 'en'}
lang_code = lang_map.get(lang_choice, 'en')

print("üì§ Upload an image with English text:")
uploaded = files.upload()

for file_name in uploaded.keys():
    image = Image.open(file_name)
    extracted_text = pytesseract.image_to_string(image, lang='eng')
    print("\nüìù Extracted Text:\n", extracted_text)

    translated_text = translate_text(extracted_text, lang_code)
    print(f"\nüî§ Translation ({lang_code}):\n", translated_text)

    braille_text = convert_to_braille(translated_text, lang_code)
    beautified_output = beautify_braille_output(braille_text)
    print("\n‚†ø Beautified Braille Output:\n")
    print(beautified_output)

    print("\nüîà Playing Audio:")
    display(speak_text(translated_text, lang_code))
```
